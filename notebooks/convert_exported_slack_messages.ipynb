```python
from pathlib import Path

import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
PATH = Path("../myslackchannel Slack export Jun 9 2019 - Sep 25 2021")
EXPORT_PATH = Path('../export')
```

```python
channel_df = pd.read_json(PATH / 'channels.json')
channel_df.to_csv(EXPORT_PATH / 'channels.csv')
channel_df.head()
```

```python
integration_log_df = pd.read_json(PATH / 'integration_logs.json')
integration_log_df.to_csv(EXPORT_PATH / 'integration_logs.csv')
integration_log_df.head()
```

```python
user_df = pd.read_json(PATH / 'users.json')
user_df.to_csv(EXPORT_PATH / 'users.csv')
user_df.head()
```

```python
channel_paths = [p for p in sorted(PATH.iterdir()) if p.is_dir()]
channel_paths
```

```python
messages = []
for channel_path in tqdm(channel_paths):
    for json_path in channel_path.iterdir():
        message_df = pd.read_json(json_path)
        message_df['channel'] = channel_path.name
        messages.append(message_df)
```

```python
message_df = pd.concat(messages).reset_index(drop=True)
message_df.to_csv(EXPORT_PATH / 'messages.csv')
message_df
```
